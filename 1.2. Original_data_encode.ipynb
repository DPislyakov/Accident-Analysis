## Разбираемся с последствиями дтп. Переклассифицируем. Пытаемся сбалансировать данные.

```python
dtp.posled.value_counts()
```

## Разбираемся с объектом дорожной сети, на котором произошло дтп. Переклассифицируем. Пытаемся сбалансировать данные.

```python
dtp.sdor.value_counts()
```

```python
dtp.loc[dtp['sdor'].str.contains('Остановка|Пешеходная зона|Автостоянка|Регулируемый пешеходный переход|Нерегулируемый пешеходный переход|Тротуар'), 'sdor'] = 'Пешеходная зона'
dtp.loc[dtp['sdor'].str.contains('Нерегулируемый перекрёсток|Нерегулируемый перекресток|Нерегулируемое пересечение с круговым движением'), 'sdor'] = 'Нерегулируемый перекрёсток'
dtp.loc[dtp['sdor'].str.contains('Регулируемый перекрёсток|Регулируемый перекресток'), 'sdor'] = 'Регулируемый перекрёсток'
dtp.loc[dtp['sdor'].str.contains('Выезд с прилегающей территории'), 'sdor'] = 'Выезд с прилегающей территории'
dtp.loc[dtp['sdor'].str.contains('ж/д'), 'sdor'] = 'ж/д переезд'
dtp.loc[dtp['sdor'].str.contains('Мост, эстакада, путепровод'), 'sdor'] = 'Мост, эстакада, путепровод'
dtp.loc[dtp['sdor'].str.contains('Внутридворовая территория'), 'sdor'] = 'Внутридворовая территория'
dtp.loc[dtp['sdor'].str.contains('нет объектов на месте ДТП'), 'sdor'] = 'Перегон (нет объектов на месте ДТП)'
dtp.loc[dtp['sdor'].str.contains('Ледовая переправа|ледовая переправа|Тоннель|АЗС|гараж|Место для перегона скота|Подход к мосту, эстакаде, путепроводу|Иное место'), 'sdor'] = 'Иное'
```

```python
dtp.sdor.value_counts()
```

```python
encoder = OrdinalEncoder()
dtp[['dtp_v', 's_pch', 'osv', 'pogoda', 'change_org_motion', 'posled', 'sdor']] = encoder.fit_transform(dtp[['dtp_v', 's_pch', 'osv', 'pogoda', 'change_org_motion', 'posled', 'sdor']])
```

```python
dtp[['dtp_v', 's_pch', 'osv', 'pogoda', 'change_org_motion', 'posled', 'sdor']] = dtp[['dtp_v', 's_pch', 'osv', 'pogoda', 'change_org_motion', 'posled', 'sdor']].astype(int)
```
