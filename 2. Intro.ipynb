## Данная глава посвящена построению прогностических моделей. Целью является проверить полноту исходного набора данных, важность отдельных атрибутов для построения моделей и исследовать поведение моделей на исходном наборе данных.

## В качестве моделей классификации были выбраны Random Forest, Support Vector Machine, KNeighbors.

# Модель Random Forest имеет подход к классификации аналогичный дереву решений, то есть входный набор данных с некоторой случаностью разделяется на более мелкие решения. Цель модели состоит в группировке результатов на основе наиболее вероятных положительных ответов. 
## Получающийся в результате разнообразный лес некоррелированных деревьев демонстрирует уменьшенную дисперсию. Следовательно, он более устойчив к изменениям данных и переносит точность прогноза на новые данные.

# Алгоритм KNN является одним из простейших методов классификации. Модель работает по принципу присваивания объекту класса, принадлежащего большинтсву его ближайших соседей в многомерном пространстве признаков. 
## Алгоритм высчитывает расстояние до каждого из объектов обучающей выборки, на которой выбирает k объектов, расстояние до которых минимально. В итоге, объекту присваивается класс, который чаще всего встречается среди k-ближайших соседей. KNN устойчив к зашумленным данным, поскольку он усредняет k-ближайших соседей.


# Алгоритм SVM находит гиперплоскость в N-мерном пространстве для разделения точек по классам. Переменная N в данном случае выступает в качестве количества классов. 
## Выбранная алгоритмом гиперплоскость напрямую влияет на точность полученных результатов. В итоге, модель SVM ищет плоскость в многомерном пространстве, которая имеет максимальное расстояние между точками данных классов. Модель SVM показывают точные результаты с минимальной вычислительной мощностью.
## В данной главе будет рассмотрена SVM  модель с разными ядрами - линейным и RBF (радиальная базисная функция). 
## Ядро rbf добавляет метод признаков близости.он позволяет получить похожий результат, как если бы добавлялись многочисленные признаки близости, без фактического их добавления.
