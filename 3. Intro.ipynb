# Данная глава посвящена построению временных рядов. Цель состоит в нахождении наиболее точной модели и прогнозировании отдельных атрибутов с помощью временных рядов. 
## Необходимо исследовать поведение исходного набора данных в результате применения определенных моделей, выявить среднюю ошибку для каждой соотвественно.

## Основная задача состоит в построении модели с точки зрения временных рядов. Под временным рядом следует понимать собранный в разные моменты времени статистический материал о значении каких-либо параметров (в простейшем случае одного) исследуемого процесса. Каждая единица статистического материала называется измерением или отсчётом. Во временном ряде для каждого отсчёта должно быть указано время измерения или номер измерения по порядку. Всякий временной ряд включает два обязательных элемента: время и конкретное значение показателя, или уровень ряда.
## Анализ временных рядов предполагает, что данные содержат систематическую составляющую (обычно включающую несколько компонент) и случайный шум (ошибку), который затрудняет обнаружение регулярных компонент. Большинство методов исследования временных рядов включает различные способы фильтрации шума, позволяющие увидеть регулярную составляющую более отчетливо. Большинство регулярных составляющих временных рядов принадлежит к двум классам: они являются либо трендом, либо сезонной составляющей. Тренд представляет собой общую систематическую линейную или нелинейную компоненту, которая может изменяться во времени. Сезонная составляющая — это периодически повторяющаяся компонента. Оба эти вида регулярных компонент часто присутствуют в ряде одновременно.

# В качестве методов прогнозирования временных рядов выбраны следующие алгоритмы:


# 1.  Авторегрессионная интегрированная скользящая средняя (ARIMA)

## Метод авторегрессионного интегрированного скользящего среднего (ARIMA) моделирует следующий шаг в последовательности как линейную функцию разностных наблюдений и остаточных ошибок на предыдущих временных шагах.
## Он сочетает в себе модели авторегрессии (AR) и скользящего среднего (MA), а также этап предварительной обработки разности, чтобы сделать последовательность стационарной, называемую интегрированием (I), то есть рассматривается ряд разностей, а не исходный ряд.
## Скользящим среднем является  семейство функций, значения которых в каждой точке определения равны некоторому среднему значению исходной функции за предыдущий период.
## Обозначение для модели включает определение порядка для моделей AR (p), I (d) и MA (q) в качестве параметров функции ARIMA, например ARIMA (p, d, q), где p - параметры авторегрессии, d - порядок разности, q - параметры скользящего среднего.

# 2. Prophet

## Данная модель прогнозирования позволяет учитывать несколько сезонных факторов. Prophet считает, что временной ряд может быть разложен на сумму тренда, сезонности, аномальных данных и ошибки.
## Подгонка модели представляет собой упражнение по подгонке кривой, поэтому она явно не учитывает структуру временной зависимости в данных. Это также позволяет проводить наблюдения с нерегулярным интервалом.
## Есть два варианта временных рядов тренда: модель насыщающего роста и кусочно-линейная модель. Модель многопериодной сезонности основана на рядах Фурье. 
## Модель Prophet вставлена в байесовскую структуру и позволяет сделать полный апостериорный вывод, чтобы включить неопределенность параметров модели в неопределенность прогноза.

# 3. LSTM ( Сеть долго-краткосрочной памяти)

## Состояние сети LSTM представлено через вектор пространства состояний. Этот метод позволяет отслеживать зависимости новых наблюдений от прошлых (даже очень далеких).
## LSTM представляют собой сложные модели, и они редко используются для прогнозирования одного временного ряда, поскольку для их оценки требуется большой объем данных. Однако они обычно используются, когда необходимы прогнозы для большого количества временных рядов
